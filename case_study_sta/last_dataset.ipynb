# Stepwise Reg.

```python
# randomise spliting
from sklearn.model_selection import train_test_split
X = df_trans[['YEAR', 'MONTH', 'QUARTER', 'QUANTITY_ORDERED',
       'PRICE_EACH', 'MSRP', 'SALES', 'STATUS', 'PRODUCT_LINE', 'STATE',
       'DEAL_SIZE']].reset_index(drop=True)
Y = df_trans[['SALES']].reset_index(drop=True)
X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.3, random_state=101)
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn import preprocessing
from itertools import combinations

def for_col_selector(df_):

    "list of combinations of columns"
    
    _lst = []
    _lst_comb = []
    for _ in range(df_.shape[1]):
        _lst.append(_)
    idx=0
    for ind in range(df_.shape[1]):
        temp = combinations(_lst, ind+1)
        for i in list(temp):
            idx = idx+1
            # print ("Combination", idx, ": ", i)
            _lst_comb.append(list(i))
    return _lst_comb


from sklearn.linear_model import LinearRegression
def lin_reg_model(x_train, y_train, x_test, y_test):

    '''error measured with MAE'''
    
    reg_mod = LinearRegression()
    reg_mod.fit(x_train, y_train)
    pred = reg_mod.predict(x_test)
    err = mean_absolute_error(y_test, pred)
    return err


def step_wise_brut(x_train, y_train, x_test, y_test):

    '''forward stepwise for all of the columns'''
    dict = {'comb': [], 'MAE': []}
    lst_col_ind = for_col_selector(x_train)
    
    for _ in range(len(lst_col_ind)):
        train_tab_x = x_train.iloc[:, lst_col_ind[_]]
        test_tab_x = x_test.iloc[:, lst_col_ind[_]]
        ress_err = lin_reg_model(train_tab_x, y_train,test_tab_x, y_test)
        print(f'The MAE in {lst_col_ind[_]} is {ress_err}')

        dict["comb"].append(lst_col_ind[_])
        dict["MAE"].append(ress_err)
    return dict

step_wise_brut(X_train,y_train, X_test, y_test)
```

```python
# non randomize spliting
X_train = X.iloc[:round(.8*len(X)),:]
X_test = X.iloc[round(.8*len(X)):,:]
y_train = Y.iloc[:round(.8*len(Y)),:]
y_test = Y.iloc[round(.8*len(Y)):,:]

step_wise_brut(X_train,y_train, X_test, y_test)
```
